## **1.2 Movimento Retilíneo Uniforme (MRU)**

### **1.2.1 Posição em função do tempo no MRU**

A velocidade média de um objeto é definida como:

$$\overrightarrow{v}_{\mathrm{med.}} = \frac{\overrightarrow{\Delta s}}{\Delta t}.$$ 

Dizemos que um movimento é uniforme, quando o deslocamento do objeto $\overrightarrow{\Delta s}$ é o mesmo em intervalos de tempo iguais. A velocidade média é sempre a mesma e não depende do intervalo escolhido para medi-la.

Em um movimento retilíneo, o objeto se movimenta sempre sobre uma mesma reta, isto quer dizer que o objeto sempre se move na mesma direção e sempre no mesmo sentido. Portanto, quando um objeto está em MRU, podemos dizer que seu vetor velocidade é constante ($\overrightarrow{v} = \mathrm{constante}$). 

```julia
using Plots
using LaTeXStrings

#definindo número de pontos dentro dos vetores posição e tempo
npoints = 301

#definindo posições na direção x do objeto
x1 = range(100, 550, length = npoints)

#fixando posição y do objeto
y1 = zeros(npoints)

#definindo instantes de tempo
tempo = range(0,150, length = npoints)

#título do eixo da posição
xlabel = "Posição em x (metros)"

#inicializando vetores utilizados para marcar posição e tempo na animação
xpace = []
ypace= []
time_ann = []

#criando animações
@gif for i = 1:npoints
    #animação mostrando o movimento do objeto
    p1 = plot([x1[i]],[y1[i]], shape = :circle, markersize = 8, ylims = (-10,10),
        xaxis = (xlabel, (0,450), 70:30:430, font(12)), 
        yaxis = (showaxis = :off), legend = :none, title = "Movimento Retilíneo Uniforme (MRU) - Objeto 1")
    #rastro do objeto em movimento
    plot!(x1[1:i],y1[1:i], lw = 3, color = :blue, linestyle = :dash, la = 0.3)
    
    #definindo os pontos que serão utilizados na marcação da distância e do tempo
    if (tempo[i] % 10 == 0)
        push!(xpace,[x1[i],x1[i]])
        push!(ypace,[-10,0])
        push!(time_ann,(x1[i],1,text(string("t = ", round(Int64, tempo[i]), "s"), pointsize = 9, color = :blue)))
    end
    
    #criando marcações na animação do movimento do objeto
    plot!(xpace,ypace, lw = 3, linestyle = :dot, color = :blue, ann = time_ann)
    
    #animação do gráfico de posição x tempo
    p2 = plot([tempo[i]],[x1[i]], shape = :circle, markersize = 8, xaxis = ("Tempo (s)",(0,110),0:10:110, font(12)),
            yaxis = ("Posição em x(m)", (0,500), 70:30:430, font(12)), label = :none, 
            title = "Posição x tempo do Objeto 1")
    plot!(tempo[1:i],x1[1:i], lw=3, color = :blue, linestyle = :dash, la = 0.3, label = "Velocidade positiva")
    
    #desenhando os gráficos mostrando movimento e posição em função do tempo
    plot(p1,p2, layout = grid(2,1, heights = [0.3,0.7]), size = (900,800))
end
```

Quando o objeto está se movendo em MRU, o objeto se move em apenas uma dimensão, e, portanto, necessitamos de apenas um número para indicar a posição do objeto sobre a reta. Sobre uma determinada reta, é possível se movimentar em dois sentidos diferentes. Sendo assim, sempre podemos escolher um dos sentidos como sendo o positivo, e o sentido oposto , portanto, será o sentido negativo. 

No caso mostrado acima, a posição do objeto, em relação a um certo ponto de referência, aumenta à medida que o tempo passa. Portanto a velocidade do objeto é positiva.

Quando a posição do objeto diminui à medida que o tempo passa, dizemos que a velocidade do objeto é negativa (ver caso abaixo).

```julia
npoints = 301

x2 = range(100, -500, length = npoints)
y2 = zeros(npoints)
#tempo2 = range(0,300, length = npoints)

xlabel = "Posição em x (metros)"

xpace = []
ypace= []
time_ann = []

@gif for i = 1:npoints
    p1 = plot([x2[i]],[y2[i]], shape = :square, markersize = 8, ylims = (-10,10),
        xaxis = (xlabel, (-400,150), -380:40:140, font(12)), color = :red, 
        yaxis = (showaxis = :off), legend = :none, title = "Movimento Retilíneo Uniforme (MRU) - Objeto 2")
    plot!(x2[1:i],y2[1:i], lw = 3, color = :red, linestyle = :dash, la = 0.3)
    
    if (tempo[i] % 20 == 0)
        push!(xpace,[x2[i],x2[i]])
        push!(ypace,[-10,0])
        push!(time_ann,(x2[i],1,text(string("t = ", round(Int64, tempo[i]), "s"), pointsize = 9, color = :red)))
        
    end
    
    plot!(xpace,ypace, lw = 3, linestyle = :dot, color = :red, ann = time_ann)
    
    p2 = plot([tempo[i]],[x2[i]], shape = :square, markersize = 8, xaxis = ("Tempo (s)",(0,120),0:20:120, font(12)),
            yaxis = ("Posição em x(m)", (-400,150), -380:40:140, font(12)), color = :red,
            label = :none, title = "Posição x tempo do Objeto 2", framestyle = :origin)
    plot!(tempo[1:i],x2[1:i], lw=3, color = :red, linestyle = :dash, la = 0.3, label = "Velocidade negativa")
    
    plot(p1,p2, layout = grid(2,1, heights = [0.3,0.7]), size = (900,800))
end
```

Se soubermos a equação da reta que representa o movimento acima, podemos descobrir a posição do objeto em qualquer instante de tempo, até mesmo para aqueles que não estão representados no gráfico.

A equação de uma reta é sempre uma função do tipo: $f(x) = ax + b$. No caso acima, temos uma reta que relaciona a posição de um objeto $s$ com um instante de tempo $t$, portanto, nesse caso teremos a posição em função do tempo: $s(t) = at + b$. Nesse caso devemos descobrir o significado físico dos coeficientes $a$ e $b$ da reta.

Substituindo $t=0$ na equação de $s(t)$, encontramos:

$$s(0) = a \cdot 0 + b = b.$$

Mas a posição no instante $t=0$ é a posição inicial do objeto $s_0$. Se $s(0) = b$, então $b$ é a posição inicial $s_0$ ($b = s_0$).

Já o coeficiente $a$ na equação da reta é chamado de coeficiente angular da reta (ou inclinação da reta). Para calcular o valor de $a$, precisamos conhecer a posição do objeto em dois instantes de tempo quaisquer. Por exemplo, se soubermos a posição inicial $s_0$, e a posição $s_1$ em um instante qualquer $t_1$, substituimos esses valores na equação de $s(t)$:  
$$ s_1 = a\cdot t_1 + s_0$$

$$ a = \frac{s_1 - s_0}{t_1 - 0}.$$

O coeficiente angular $a$ é o deslocamento $\Delta s = s_1 - s_0$ dividido pelo intervalo de tempo $\Delta t = t_1 - 0$, ou seja, $a$ é igual à velocidade média no intervalo entre $t = 0$ e $t_1$. Como no caso do MRU, a velocidade do objeto $v$ é sempre igual à sua velocidade média: $a = v$. Assim sendo, podemos reescrever a equação da reta, que descreve um objeto em MRU, como:

<div class="alert alert-block alert-warning">
<b>Equação da posição em função do tempo (MRU):</b>
$$ s(t) = v\cdot t + s_0.$$
</div>

O deslocamento em função do tempo $\Delta s(t) = s(t) - s_0$ pode ser encontrado a partir da equação acima:

<div class="alert alert-block alert-warning">
<b>Equação do deslocamento em função do tempo (MRU):</b>
$$ \Delta s(t) = v\cdot t.$$
</div>


**A inclinação da reta, em um gráfico de posição por tempo, representa a velocidade $v$ do objeto.** Isso confirma o que vimos anteriormente, quando percebemos que quanto maior a inclinação da reta num gráfico de posição por tempo, maior a velocidade do objeto!

```julia
# selecionar pontos utilizados para calcular deslocamentos, intervalo de tempo e velocidade
p1 = 22
p2 = 219

#criando anotações com as posições cartesianas dos dois pontos escolhidos p1 e p2
p1str = string("(", round(tempo[p1], digits = 1), " s, ", round(x2[p1], digits = 1), " m )")
p1str = replace(p1str, "." => ",")
p2str = string("(", round(tempo[p2], digits = 2), " s, ", round(x2[p2], digits = 1), " m )")
p2str = replace(p2str, "." => ",")
annp1 = [(tempo[p1], x2[p1] + 30, text(p1str, pointsize = 10, halign = :center, color = :darkblue)),
   (tempo[p2], x2[p2] + 30, text(p2str, pointsize = 10, halign = :center, color = :darkblue))]

#criando anotações com os deslocamentos e intervalos de tempo
dxstr = latexstring("\\Delta x = ", round((x2[p2]-x2[p1]), digits = 1), "\\, \\mathrm{m}")
dxstr = replace(dxstr, "." => ",\\!")
pxdx = tempo[p2] - 2
pydx = (x2[p2]+x2[p1])/2

dtstr = latexstring("\\Delta t = ", round((tempo[p2]-tempo[p1]), digits = 1), "\\, \\mathrm{s}")
dtstr = replace(dtstr, "." => ",\\!")
pxdt = (tempo[p2] + tempo[p1])/2
pydt = x2[p1] + 30

anndel = [(pxdx,pydx, text(dxstr, pointsize = 12, halign = :right, color = :darkblue)),
  (pxdt,pydt, text(dtstr, pointsize = 12, halign = :center, color = :darkblue))]

#criando anotações com a equação da velocidade e o resultado obtido através dos pontos p1 e p2 escolhidos
v2 = (x2[p2]-x2[p1])/(tempo[p2]-tempo[p1])
velstr = latexstring("v = \\frac{\\Delta x}{\\Delta t} = \\frac{", (x2[p2] - x2[p1]), "\\, \\mathrm{m}}{",
   (tempo[p2]-tempo[p1]), "\\, \\mathrm{s}} = ", v2, "\\, \\mathrm{m/s}")
velstr = replace(velstr, "." => ",\\!")

annvel = [(20,-400, text(velstr, pointsize = 12, halign = :left, color = :darkblue))]


plot(tempo, x2, color = :red, label = :none, xaxis = ("tempo (s)", (0,150), 0:10:150),
   yaxis = ("Posição em x (m)", (-500,200), -500:50:200), lw = 3, 
   title = "Posição x tempo \n (Objeto 2)", size = (900,500), legend = :none, framestyle = :origin)
#plot!([0,150],[0,0], lw = 1, color = :black)

scatter!([tempo[p1]],[x2[p1]], color= :red, markersize = 8, label = :none, ann = annp1)
plot!([0,tempo[p1]],[x2[p1],x2[p1]], color = :red, lw = 2, ls = :dash, la = 0.3)
plot!([tempo[p1],tempo[p1]],[0,x2[p1]], color = :red, lw = 2, ls = :dash, la = 0.3)

scatter!([tempo[p2]],[x2[p2]], color = :red, markersize = 8, label = :none)
plot!([0,tempo[p2]],[x2[p2],x2[p2]], color = :red, lw = 2, ls = :dash, la = 0.3)
plot!([tempo[p2],tempo[p2]],[0,x2[p2]], color = :red, lw = 2, ls = :dash, la = 0.3, ann = annvel)

plot!([tempo[p1],tempo[p2]],[x2[p1],x2[p1]], color = :blue, lw = 2, ls = :dash, la = 0.7)
plot!([tempo[p2],tempo[p2]],[x2[p1],x2[p2]], color = :blue, lw = 2, ls = :dash, la = 0.7, ann = anndel)
```
