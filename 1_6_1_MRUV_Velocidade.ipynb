## 1.6  Movimento Retilíneo Uniformemente Variado (MRUV)

### 1.6.1  Velocidade em função do tempo no MRUV

O tipo de movimento mais simples possível é o movimento onde a velocidade é constante: movimento retilíneo uniforme. O próximo passo é entender o que acontece com o movimento quando a velocidade não é mais constante.

Por enquanto, o movimento continua sendo retilíneo: a direção do movimento não muda. Mas vamos considerar que o módulo e o sentido da velocidade podem variar de forma uniforme. Esse tipo de movimento é chamado de movimento retilíneo uniformemente variado (MRUV).

Quando dizemos que a velocidade varia de maneira uniforme, isso significa que a velocidade sempre varia a mesma quantidade $\Delta v$ em qualquer intervalo de tempo igual a $\Delta t$. O mesmo tipo de comportamento que ocorria com a posição no MRU.

A figura abaixo mostra um exemplo de velocidade que varia uniformemente com o tempo no MRUV.

```julia
using Plots
using LaTeXStrings
using Printf

#definindo a função v(t): a é a aceleração, v0 é a velocidade inicial, t é o instante de tempo
function vel(a::Real, v0::Real, t::Real)
    return a*t + v0
end

#definindo características padrões para os gráficos
default(size = [800,600], grid = true, markersize = 8, framestyle = :origin, linewidth = 3, legend = :false)

#definindo vetor com os valores de t utilizados para desenhar o gráfico
t = 0:1:20
tmax = maximum(t)

#definindo vetor com as velocidades utilizadas para desenhar o gráfico
a = 2
v0 = 3
vx = vel.(a,v0,t)
vxmax = maximum(vx)

#desenhando o gráfico
plot(t,vx, xaxis = ("tempo (s)", (0, tmax), 0:2:tmax, font(10)),
   yaxis = ("velocidade (m/s)", (0,vxmax), v0:2*a:vxmax, font(10)), title = "velocidade x tempo")
```

Novamente, temos uma reta representando uma grandeza em função do tempo. Toda vez que uma grandeza varia uniformemente em função de outra, o gráfico que representa essa variação é uma reta.

Vamos aplicar o mesmo tipo de tratamento matemático que aplicamos ao gráfico de posição x tempo no MRU, mas no caso do MRUV, aplicando ao gráfico de velocidade x tempo. Lembrando que a equação de um gráfico linear como o mostrado acima é do tipo: $f(x) = a\cdot x + b$. Como temos a velocidade $v$ em função do tempo $t$, a equação será escrita como:

$$ v(t) = a\cdot t + b.$$

O coeficiente $a$ é o coeficiente angular ou inclinação da reta que representa a função $v(t)$. O coeficiente $b$ é igual à velocidade do objeto no instante $t=0$, ou seja, $b$ é a velocidade inicial $v_0$ do objeto. 

Para entender o significado da inclinação do gráfico nesse caso, vamos comparar algumas velocidades de objetos em MRUV.

```julia
#desenhando gráficos de velocidade x tempo no MRUV com várias inclinações diferentes.

plot(t,vx, xaxis = ("tempo (s)", (0, tmax), 0:2:tmax, font(10)),
    yaxis = ("velocidade (m/s)", (-vxmax,vxmax), -vxmax:2*a:vxmax, font(10)), title = "velocidade x tempo",
    label = latexstring("{\\rm Objeto ~ 1:} ~ v_1(t)"), legend = :bottomleft, legendfontsize = 10)
plot!(t,vel.(2*a,v0,t), label = latexstring("{\\rm Objeto ~ 2:} ~v_2(t)"))
plot!(t,vel.(a/2,v0,t), label = latexstring("{\\rm Objeto ~ 3:} ~v_3(t)"))
plot!([0,tmax],[v0,v0], label = latexstring("{\\rm Objeto ~ 4:} ~v_4(t)"))
plot!(t,vel.(-a,v0,t), label = latexstring("{\\rm Objeto ~ 5:} ~v_5(t)"))
plot!(t,vel.(-1.5*a,v0,t), label = latexstring("{\\rm Objeto ~ 6:} ~v_6(t)"))
plot!([8,8],[-vxmax,vxmax], lw = 1, ls = :dash, color = :black, label = :none)
```

Todos os objetos têm a mesma velocidade inicial $v_0$ para simplificar a análise da inclinação. A velocidade do objeto 2 é a que crescer mais rápida entre todos os objetos representados. Como a velocidade cresce a medida que o tempo passa, dizemos que a reta que a representa tem inclinação positiva, portanto, a reta que representa a velocidade do objeto 2 tem a maior inclinação. Já a velocidade do objeto 6 é a que decresce mais rapidamente entre todos os objetos. Quando a velocidade diminui a medida que o tempo passa, dizemos que a reta que a representa tem inclinação negativa, portanto, a reta que representa a velocidade do objeto 6, tem a menor inclinação. A velocidade do objeto 4 é constante, portanto, a reta que a representa tem inclinação nula, ou seja, o objeto está em MRU.

A grandeza que representa o quanto a velocidade de um objeto varia com o tempo é chamada de *aceleração*, e também é uma grandeza vetorial representada como $\vec{a}$. Lembrando que, enquanto estivermos analisando movimentos retilíneos, onde não há mudança de direção no movimento, não precisamos nos preocupar com a notação e nem a realidade vetorial das grandezas, por isso vamos representar a aceleração apenas como $a$.

Assim sendo, a inclinação do gráfico é igual a aceleração $a$ do movimento, e a equação que representa a velocidade em função do tempo pode ser escrita como:

$$v(t) = a\cdot t + v_0.$$

```julia
#desenhando gráficos de velocidade x tempo no MRUV com várias inclinações diferentes.

plot(t,vx, xaxis = ("tempo (s)", (0, tmax), 0:2:tmax, font(10)),
    yaxis = ("velocidade (m/s)", (-vxmax,vxmax), -vxmax:2*a:vxmax, font(10)), title = "velocidade x tempo",
    label = latexstring("{\\rm Objeto ~ 1:} ~ a_1 > 0"), legend = :bottomleft, legendfontsize = 10)
plot!(t,vel.(2*a,v0,t), label = latexstring("{\\rm Objeto ~ 2:} ~ a_2 > a_1 > 0"))
plot!(t,vel.(a/2,v0,t), label = latexstring("{\\rm Objeto ~ 3:} ~ a_1 > a_3 > 0"))
plot!([0,tmax],[v0,v0], label = latexstring("{\\rm Objeto ~ 4:} ~ a_4 = 0"))
plot!(t,vel.(-a,v0,t), label = latexstring("{\\rm Objeto ~ 5:} ~ a_5 < 0"))
plot!(t,vel.(-1.5*a,v0,t), label = latexstring("{\\rm Objeto ~ 6:} ~ a_6 < a_5 < 0"))
```

### 1.6.2 Aceleração ($\vec{a}$) no MRUV

**No MRUV, a velocidade do objeto varia, mas com aceleração constante!**

Para se calcular a aceleração do objeto (*inclinação da reta*) a partir do gráfico de velocidade em função do tempo temos que:

* escolher dois instantes de tempo $t_1$ e $t_2$ quaisquer, nos quais sejam conhecidas as velocidades do objeto $v_1$ e $v_2$ nos instantes respectivos;
* calcular a variação da velocidade $\Delta v = v_2 - v_1$, e o intervalo de tempo no qual ocorreu essa variação de velocidade $\Delta t = t_2 - t_1$;
* por fim, a inclinação é dada por:

$$ a = \frac{\Delta v}{\Delta t}.$$

```julia
#escolhendo instantes de tempo t1 e t2 que serão utilizado no cálculo da aceleração
t1 = 6; t2 = 17;
Δt = t2 - t1

#calculando as velocidades nos respectivos instantes de tempo
v11 = vel(a,v0,t1)
v12 = vel(a,v0,t2)
Δv = v12 - v11

#calculando a aceleração do objeto
a1 = Δv/Δt

#criando strings para anotações
#string para Δt
dtstr = @sprintf("%3.1f", Δt)
dttext = latexstring("\\Delta t = ", dtstr, "\\,{\\rm s}")
dttext = replace(dttext, "." => ",\\!")
#definindo posição da anotação sobre Δt
dttext_pos = [(t2+t1)/2, v11 - 1]

#strings para Δv
dvstr = @sprintf("%3.1f", Δv)
dvtext = latexstring("\\Delta v = ", dvstr, "\\,{\\rm m/s}")
dvtext = replace(dvtext, "." => ",\\!")
#definindo posição da anotação sobre Δv
dvtext_pos = [t2 + 0.4, (v11+v12)/2]

#string para a
a1str = @sprintf("%3.2f", a1)
a1text = latexstring("a = \\frac{\\Delta v}{\\Delta t} = \\frac{", dvstr, "\\,{\\rm m/s}}{", 
    dtstr, "\\,{\\rm s}} =", a1str, "\\,{\\rm m/s^2}")
a1text = replace(a1text, "." => ",\\!")
#definindo posição da anotação sobre a
a1text_pos = [3, vxmax - 2]

#criando vetor de anotações
anots = [(dttext_pos[1], dttext_pos[2], text(dttext, halign = :center, pointsize = 12, color = :blue)),
    (dvtext_pos[1], dvtext_pos[2], text(dvtext, valign = :center, pointsize = 12, color = :blue, rotation = 90)),
    (a1text_pos[1], a1text_pos[2], text(a1text, halign = :left, pointsize = 12, color = :blue))]

#desenhando gráfico de v x t
plot(t,vx, xaxis = ("tempo (s)", (0, tmax), [0,t1,t2, tmax], font(10)),
    yaxis = ("velocidade (m/s)", (0,vxmax), [0,v0,v11,v12,vxmax], font(10)), title = "velocidade x tempo")

#desenhando pontos escolhidos no gráfico
scatter!([t1,t2],[v11,v12], marker = :circle, color = :red, alpha = 0.5)

#marcando ponto t1, v1 no gráfico
plot!([0,t1],[v11,v11], color = :red, alpha = 0.3, ls = :dash)
plot!([t1,t1],[0,v11], color = :red, alpha = 0.3, ls = :dash)

#marcando ponto t2,v2 no gráfico
plot!([0,t2],[v12,v12], color = :red, alpha = 0.3, ls = :dash)
plot!([t2,t2],[0,v12], color = :red, alpha = 0.3, ls = :dash)

#marcando comprimentos de Δt e Δv no gráfico
plot!([t1,t2],[v11,v11], color = :dodgerblue, alpha = 0.5, ls = :dash)
plot!([t2,t2],[v12,v11], color = :dodgerblue, alpha = 0.5, ls = :dash)

#anotando valores no gráfico
annotate!(anots)
```

Note que a aceleração $a$ é a mesma para quaisquer instantes $t_1$ e $t_2$ que forem escolhidos.

A equação acima é utilizada, na verdade, no cálculo da aceleração média $a_{\rm med}$ de um objeto durante um intervalo de tempo $\Delta t$. Entretanto, quando um objeto está em MRUV, sua aceleração é constante e, assim sendo, será igual à sua aceleração média durante todo o movimento.

Já no caso de movimentos não-retilíneos, onde a direção do objeto também varia com o tempo, deve-se levar em consideração a identidade vetorial das grandezas e, então, a aceleração média é:

$$\overrightarrow{a}_{\rm \! med} = \frac{\overrightarrow{\Delta v}}{\Delta t},$$

e, dificilmente, esta será igual a aceleração no instante $t_1$ ($\overrightarrow{a}_{\! 1}$) , ou a aceleração no instante $t_2$ ($\overrightarrow{a}_{\! 2}$).

A próxima figura mostra a representação gráfica da aceleração em função do tempo no MRUV.

```julia
ytitle = string("aceleração (m/s²)")
#ytitle = latexstring("\\textrm{acelera\\c{c}\\~{a}o} \\, ({\\rm m/s^2})")

plot([0,tmax],[a1,a1], xaxis = ("tempo (s)", (0, tmax), [0,20], font(10)), 
    yaxis = (ytitle, (0, 2*a1), [0,a1,2*a1], font(10)), color = :forestgreen,
    title = "aceleração x tempo")
```

A área entre a reta que representa a aceleração em função do tempo e o eixo do tempo, limitada por dois instantes de tempo $t_1$ e $t_2$ é igual à variação da velocidade no intervalo de tempo $\Delta t = t_2 - t_1$. No caso do MRUV, a figura formada é um retângulo.

```julia
#cálculo da área do retângulo
dv = a1 * Δt

#gerando string com valor da aceleração - altura do retângulo
aceltext = latexstring("a = ", a1str, "\\,{\\rm m/s^2}")
aceltext = replace(aceltext, "." => ",\\!")

#gerando string com o cálculo da área do retângulo
areastr = @sprintf("%3.1f", dv)
areatext = latexstring("{\\rm \\acute{a}rea} = ", a1str, "\\,{\\rm \\frac{m}{s^2}} \\cdot ", dtstr, "\\,{\\rm s} = ",
    areastr, "\\,{\\rm \\frac{m}{s}}")
areatext = replace(areatext, "." => ",\\!")

#gerando anotações no gráfico
anots = [(dttext_pos[1], a1+0.1, text(dttext, halign = :center, pointsize=14, color = :green)),
    (t1 - 0.5, a1/2, text(aceltext, valign = :center, pointsize = 14, color = :green, rotation = 90)),
    (2, 3.6, text(areatext, halign = :left, pointsize = 14, color = :green))]

#título do eixo y
ytitle = string("aceleração (m/s²)")
#desenhando aceleração em função do tempo
plot([0,tmax],[a1,a1], xaxis = ("tempo (s)", (0, tmax), [0,t1,t2,20], font(10)), 
    yaxis = (ytitle, (0, 2*a1), [0,a1,2*a1], font(10)), color = :forestgreen,
    title = "aceleração x tempo", alpha = 0.5)

#desenhando aceleração entre t1 e t2
plot!([t1,t2],[a1,a1], color = :forestgreen, fill = (0, :palegreen), fillalpha = 0.5)

#marcando instantes t1 e t2 no gráfico
scatter!([t1,t2],[a1,a1], marker = :circle, color = :red, alpha = 0.3)
plot!([t1,t1],[0,a1], ls = :dash, color = :forestgreen)
plot!([t2,t2],[0,a1], ls = :dash, color = :forestgreen)

#escrevendo anotações
annotate!(anots)
```

A variação da velocidade encontrada pela área no gráfico de aceleração em função do tempo, é exatamente igual a variação da velocidade obtida pela análise do gráfico de velocidade em função do tempo, no mesmo intervalo observado.

A equação do cálculo da aceleração média:

$$ a_{\rm med} = \frac{\Delta v}{\Delta t},$$ 

pode ser reescrita da seguinte forma:

$$ \Delta v = a_{\rm med} \cdot \Delta t,$$

que é igual a área representada acima. Vale lembrar que **essa equação é válida apenas para MRUV**. Se o objeto não tiver aceleração constante a expressão para se calcular a variação da velocidade (*área*) será diferente.

Apesar de ser possível descobrir a variação da velocidade $\Delta v$ em um intervalo $\Delta t$ através da análise do gráfico da aceleração em função do tempo, apenas com essas informações ainda não é possível descobrir a equação de $v(t)$. Também não é possível obter a informação necessária para se obtê-la apenas com a análise do gráfico de aceleração em função do tempo. Para isso, é necessário se conhecer a velocidade do objeto em um instante apenas, qualquer instante. Sem essa informação, existem infinitas equações possíveis com a mesma aceleração. 

Vimos algo semelhante quando falamos sobre a velocidade no MRU, com alguns exemplos de como se obter a equação de posição em função do tempo, dada a posição em um instante qualquer. O problema matemático aqui é exatamente igual, mudamos apenas as grandezas: velocidade e posição no MRU por aceleração e velocidade no MRUV.
