## 1.3 Movimento Retilíneo Uniforme (MRU)

## 1.3.1 Posição em função do tempo

### A velocidade média de um objeto é definida como:

### $$\overrightarrow{v}_{\mathrm{med.}} = \frac{\overrightarrow{\Delta s}}{\Delta t}.$$ 

### Dizemos que um movimento é uniforme, quando o deslocamento do objeto $\overrightarrow{\Delta s}$ é sempre o mesmo para qualquer intervalo de tempo que seja igual a um certo $\Delta t$, ou seja, o módulo da velocidade média é constante durante todo o movimento.

### Em um movimento retilíneo, o objeto se movimenta sempre sobre uma mesma reta, isto quer dizer que o objeto sempre se move na mesma direção e sempre no mesmo sentido. Portanto, quando um objeto está em MRU, podemos dizer que seu vetor velocidade é constante ($\overrightarrow{v} = \mathrm{constante}$). 

### <font color = red>Observação: Se o vetor velocidade é constante, a velocidade será sempre igual à velocidade média!</font>

```julia
using Plots
using LaTeXStrings

npoints = 301

x1 = range(100, 550, length = npoints)
y1 = zeros(npoints)
tempo = range(0,150, length = npoints)

xlabel = "Posição em x (metros)"

xpace = []
ypace= []
time_ann = []

@gif for i = 1:npoints
    p1 = plot([x1[i]],[y1[i]], shape = :circle, markersize = 10, ylims = (-10,10),
        xaxis = (xlabel, (0,450), 70:30:430, font(12)), 
        yaxis = (showaxis = :off), legend = :none, title = "Movimento Retilíneo Uniforme (MRU) - Objeto 1")
    plot!(x1[1:i],y1[1:i], lw = 3, color = :blue, linestyle = :dash, la = 0.3)
    
    if (tempo[i] % 10 == 0)
        push!(xpace,[x1[i],x1[i]])
        push!(ypace,[-10,0])
        push!(time_ann,(x1[i],1,text(string("t = ", round(Int64, tempo[i]), "s"), pointsize = 9, color = :blue)))
        
    end
    
    plot!(xpace,ypace, lw = 3, linestyle = :dot, color = :blue, ann = time_ann)
    
    p2 = plot([tempo[i]],[x1[i]], shape = :circle, markersize = 10, xaxis = ("Tempo (s)",(0,110),0:10:110, font(12)),
            yaxis = ("Posição em x(m)", (0,500), 70:30:430, font(12)), label = :none, 
            title = "Posição x tempo do Objeto 1")
    plot!(tempo[1:i],x1[1:i], lw=3, color = :blue, linestyle = :dash, la = 0.3, label = "Velocidade positiva")
    
    plot(p1,p2, layout = (2,1), size = (950,1000))
end
```

### Quando o objeto está se movendo em MRU, o objeto se move em apenas uma dimensão, e, portanto, necessitamos de apenas um número para indicar a posição do objeto sobre a reta. 

### Sobre uma determinada reta, pode-se movimentar em dois sentidos diferentes. Sendo assim, sempre podemos escolher um dos sentidos como sendo o positivo, e o sentido oposto , portanto, será o sentido negativo. 

### No caso mostrado acima, a posição do objeto, em relação a um certo ponto de referência, aumenta à medida que o tempo passa. Portanto a velocidade do objeto é positiva.

### Quando a posição do objeto diminui à medida que o tempo passa, dizemos que a velocidade do objeto é negativa (ver caso abaixo).

```julia
npoints = 301

x2 = range(100, -500, length = npoints)
y2 = zeros(npoints)
#tempo2 = range(0,300, length = npoints)

xlabel = "Posição em x (metros)"

xpace = []
ypace= []
time_ann = []

@gif for i = 1:npoints
    p1 = plot([x2[i]],[y2[i]], shape = :square, markersize = 10, ylims = (-10,10),
        xaxis = (xlabel, (-400,150), -380:40:140, font(12)), color = :red, 
        yaxis = (showaxis = :off), legend = :none, title = "Movimento Retilíneo Uniforme (MRU) - Objeto 2")
    plot!(x2[1:i],y2[1:i], lw = 3, color = :red, linestyle = :dash, la = 0.3)
    
    if (tempo[i] % 20 == 0)
        push!(xpace,[x2[i],x2[i]])
        push!(ypace,[-10,0])
        push!(time_ann,(x2[i],1,text(string("t = ", round(Int64, tempo[i]), "s"), pointsize = 9, color = :red)))
        
    end
    
    plot!(xpace,ypace, lw = 3, linestyle = :dot, color = :red, ann = time_ann)
    
    p2 = plot([tempo[i]],[x2[i]], shape = :square, markersize = 10, xaxis = ("Tempo (s)",(0,120),0:20:120, font(12)),
            yaxis = ("Posição em x(m)", (-400,150), -380:40:140, font(12)), color = :red,
            label = :none, title = "Posição x tempo do Objeto 2", framestyle = :origin)
    plot!(tempo[1:i],x2[1:i], lw=3, color = :red, linestyle = :dash, la = 0.3, label = "Velocidade negativa")
    
    plot(p1,p2, layout = (2,1), size = (950,1000))
end
```

### Como podemos ver, no caso do MRU, independente dos pontos que escolhemos, encontramos sempre a mesma velocidade. (Este é o comportamento esperado, já que a velocidade é constante!)

### Se soubermos a equação da reta que representa o movimento acima, podemos descobrir a posição do objeto em qualquer instante de tempo, até mesmo para aqueles que não estão representados no gráfico!

### A equação de uma reta é sempre do tipo: $f(x) = ax + b$. No caso de uma reta que relaciona a posição de um objeto $s$ com o tempo $t$, a função será do tipo: $s(t) = at + b$.

### Quando utilizamos que $t=0$, a equação do lado fica $s(0) = a\cdot 0 + b = b$. Isso quer dizer que o coeficiente $b$ representa a posição do objeto no instante $t=0$, ou seja, é a posição inicial $s_0$ do objeto.

### O coeficiente $a$ na equação da reta é chamado de coeficiente angular da reta (ou inclinação da reta). Podemos calculá-lo se soubermos a posição inicial do objeto $s_0$ e a posição $s(t_1)$ em um instante qualquer $t_1$, e substituindo esses valores na equação:

### $$ s(t_1) = a\cdot t_1 + s_0$$

### $$ a = \frac{s(t_1) - s_0}{t_1 - 0},$$

### ou seja, a representa a relação entre o deslocamento do objeto entre o instante inicial $t=0$ e o instante $t_1$ e o intervalo de tempo que passou entre $t_1$ e $t = 0$. Isso quer dizer que $a$ na verdade é a velocidade $v$ do objeto. Assim sendo, podemos reescrever a equação da reta, que descreve um objeto em MRU, como:

### $$ s(t) = v\cdot t + s_0.$$

### <font color = red>A inclinação da reta, em um gráfico de posição por tempo, representa a velocidade $v$ do objeto.</font> Como vimos anteriormente, quanto maior a inclinação da reta num gráfico de posição por tempo, maior a velocidade do objeto!
