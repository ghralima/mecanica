# 1. CINEMÁTICA

## 1.1 Posição ($\vec{s}$)

A posição de um objeto representa a maneira mais simples de se chegar a este objeto a partir de um determinado ponto de referência. Por exemplo, imagine que nosso ponto de referência seja o ponto O, e todas as posições são medidas a partir deste ponto O. Assim sendo, determina-se o ponto O como sendo o centro de coordenadas do sistema:

$$ \vec{s_O} = (0,0).$$

A letra $s$ será utilizada para representar a posição dos objetos, e $\vec{s_O}$ é a posição do ponto O.

Abaixo estão representados alguns pontos próximos ao ponto O.

```python
%matplotlib inline

#carregando bibliotecas utilizadas
import matplotlib
matplotlib.rcParams['text.usetex'] = True
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams["figure.figsize"] = (20,15)

#determinamos as posições de cada um dos pontos que queremos plotar.
x0,y0 = 0,0
xA,yA = 5,7
xB,yB = -2,-6
xC,yC = -4,8

#desenhando os pontos no gráfico
fig, ax = plt.subplots()

ax.plot(x0,y0, marker = 'o', markersize = 20)
ax.text(x0-0.6,y0+0.5,'ponto O', fontsize = 20)
ax.plot(xA,yA, marker = '*', markersize = 20)
ax.text(xA-0.6,yA+0.5,'ponto A', fontsize = 20)
ax.plot(xB,yB, marker = 's', markersize = 20)
ax.text(xB-0.6,yB+0.5,'ponto B', fontsize = 20)
ax.plot(xC,yC, marker = 'P', markersize = 20)
ax.text(xC-0.6,yC+0.5,'ponto C', fontsize = 20)
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.tick_params(labelsize = 20)
ax.set_xticks(np.arange(-10,10,1))
ax.set_yticks(np.arange(-10,10,1))
ax.grid(axis = 'both')
ax.set_xlabel('Posição em X (metros)', fontsize = 20)
ax.set_ylabel('Posição em Y (metros)', fontsize = 20)

plt.show()
```

No sistema de coordenadas acima - centrado no ponto O - todas as distâncias horizontais e verticais são medidas a partir do ponto O. Posições horizontais a direita do ponto O são posições positivas, quanto mais a direta maior será a posição, posições a esquerda do ponto O são negativas, quanto mais a esquerda mais negativa será a posição. O mesmo acontece com a direção vertical, posições acima de O são positivas e posições abaixo de O são negativas. Não há problema algum em se adotar definições diferentes - por exemplo, pode-se adotar que o positivo é para esquerda e o negativo para a direita.

Existe infinitas maneiras de chegar aos pontos A, B e C partindo do ponto O, mas a maneira mais simples é ao longo de segmentos de reta que saem de O e chegam aos pontos A, B e C. Lembrando que dois pontos definem uma reta única. Esses segmentos de retas podem ser vistos na próxima figura, mas como esses segmentos de reta podem ser descritos?

```python
# desenhando os pontos no gráfico
fig, ax = plt.subplots()
ax.plot(x0,y0,marker = 'o', markersize = 20)
ax.text(x0-0.6,y0+0.5,'ponto O', fontsize = 20)
ax.plot(xA,yA, marker = '*', markersize = 20)
ax.text(xA-0.6,yA+0.5,'ponto A', fontsize = 20)
ax.plot(xB,yB, marker = 's', markersize = 20)
ax.text(xB-0.6,yB+0.5,'ponto B', fontsize = 20)
ax.plot(xC,yC, marker = 'P', markersize = 20)
ax.text(xC-0.6,yC+0.5,'ponto C', fontsize = 20)

# desenhando as setas partindo de O e chegando aos outros pontos
ax.arrow(x0,y0,xA,0, width = 0.1, length_includes_head = True, edgecolor = 'darkorange', fc = 'None')
ax.arrow(xA,0,0,yA, width = 0.1, length_includes_head = True, edgecolor = 'darkorange', fc = 'None')
ax.arrow(x0,y0,xA,yA, width = 0.15, length_includes_head = True, color = 'orange')
ax.arrow(x0,y0,xB,0, width = 0.1, length_includes_head = True, edgecolor = 'green', fc = 'None')
ax.arrow(xB,0,0,yB, width = 0.1, length_includes_head = True, edgecolor = 'green', fc = 'None')
ax.arrow(x0,y0,xB,yB, width = 0.15, length_includes_head = True, color = 'lightgreen')
ax.arrow(x0,y0,0,yC, width = 0.1, length_includes_head = True, edgecolor = 'red', fc = 'None')
ax.arrow(0,yC,xC,0, width = 0.1, length_includes_head = True, edgecolor = 'red', fc = 'None')
ax.arrow(x0,y0,xC,yC, width = 0.15, length_includes_head = True, color = 'pink')

# definindo os eixos do  gráfico
ax.set_xlim([-10,10])
ax.set_ylim([-10,10])
ax.tick_params(labelsize = 20)
ax.set_xticks(np.arange(-10,10,1))
ax.set_yticks(np.arange(-10,10,1))
ax.grid(axis = 'both')
ax.set_xlabel('Posição em X (metros)', fontsize = 20)
ax.set_ylabel('Posição em Y (metros)', fontsize = 20)

#criando e mostrando strings descrevendo a posição de A
x_dir = 'direita' if xA > 0 else 'esquerda'
y_dir = 'cima' if yA > 0 else 'baixo'

print('Para chegar em A, a partir do ponto de referência, deve-se andar {0} m para a {1} e depois {2} m para {3}!\n'.format(xA,x_dir,yA,y_dir))

print('O ponto A está na posição: ({0} m, {1} m).\n'.format(xA,yA))

#criando e mostrando strings descrevendo a posição de B
x_dir = 'direita' if xB > 0 else 'esquerda'
y_dir = 'cima' if yB > 0 else 'baixo'

print('Para chegar em B, a partir do ponto de referência, deve-se andar {0} m para a {1} e depois {2} m para {3}!\n'.format(xB,x_dir,yB,y_dir))

print('O ponto B está na posição: ({0} m, {1} m).\n'.format(xB,yB))

#criando e mostrando strings descrevendo a posição de C
x_dir = 'direita' if xC > 0 else 'esquerda'
y_dir = 'cima' if yC > 0 else 'baixo'
  
print('Para chegar em C, a partir do ponto de referência, deve-se andar {0} m para a {1} e depois {2} m para {3}!\n'.format(xC,x_dir,yC,y_dir))

print('O ponto C está na posição: ({0} m, {1} m).\n'.format(xC,yC))

#desenhando gráficos
plt.show()
#ax.arrow(x0,y0,xA,yA, width = 0.05, length_includes_head = True)
```

As setas que saem do ponto de referência e chegam aos pontos A, B e C representam as posições desses pontos em relação ao ponto de referência! Essas setas são chamadas de vetores posição $\vec{s}$. Cada uma desses vetores posição foi definido com a ajuda de dois valores: a posição horizontal e a posição vertical do ponto. Existe outras maneiras de definir o vetor posição, por exemplo, pode-se usar o ângulo que o segmento de reta que chega ao ponto A faz com a reta horizontal, e qual a distância que deve-se mover ao longo desse segmento de reta (dois valores diferentes).

Mas o que acontece com o vetor posição dos objetos se o ponto de referência é trocado?

```python
#determinando o novo ponto de referência
xR, yR = xB, yB

#desenhando os pontos na figura
fig, ax = plt.subplots()
ax.plot(x0-xR,y0-yR,marker = 'o', markersize = 20)
ax.text(x0-xR-0.1,y0-yR+0.5,(r'$\overrightarrow{{s_O}} = ({0}\,m, {1}\,m)$'.format(x0-xR,y0-yR)), 
        fontsize = 25, color = 'blue')

ax.plot(xA-xR,yA-yR, marker = '*', markersize = 20)
ax.text(xA-xR-0.1,yA-yR+0.5,r'$\overrightarrow{{s_A}} = ({0}\,m, {1}\,m)$'.format(xA-xR,yA-yR), 
        fontsize = 25, color = 'darkorange')

ax.plot(xB-xR,yB-yR, marker = 's', markersize = 20)
ax.text(xB-xR-0.1,yB-yR+0.5,r'$\overrightarrow{{s_B}} = ({0}\,m,{1}\,m)$'.format(xB-xR,yB-yR), 
        fontsize = 25, color = 'green')

ax.plot(xC-xR,yC-yR, marker = 'P', markersize = 20)
ax.text(xC-xR-0.1,yC-yR+0.5,r'$\overrightarrow{{s_C}} = ({0}\,m,{1}\,m)$'.format(xC-xR,yC-yR), 
        fontsize = 25, color = 'red')

#desenhando setas partindo do novo ponto de referência e chegando aos outros pontos
ax.arrow(0,0,xA-xR,yA-yR, width = 0.1, length_includes_head = True, color = 'orange')
ax.arrow(0,0,xB-xR,yB-yR, width = 0.1, length_includes_head = True, color = 'lightgreen')
ax.arrow(0,0,x0-xR,y0-yR, width = 0.1, length_includes_head = True, color = 'skyblue')
ax.arrow(0,0,xC-xR,yC-yR, width = 0.1, length_includes_head = True, color = 'pink')

#definindo eixos dos gráficos
ax.set_xlim([-10-xR,10-xR])
ax.set_ylim([-10-yR,10-yR])
ax.tick_params(labelsize = 20)
ax.set_xticks(np.arange(-10-xR,10-xR,1))
ax.set_yticks(np.arange(-10-yR,10-yR,1))
ax.grid(axis = 'both')
ax.set_xlabel('Posição em X (metros)', fontsize = 20)
ax.set_ylabel('Posição em Y (metros)', fontsize = 20)

#posições de O, A, B e C em relação ao ponto de referência R
x0R, y0R = x0-xR, y0-yR
xAR, yAR = xA-xR, yA-yR
xBR, yBR = xB-xR, yB-yR
xCR, yCR = xC-xR, yC-yR

# strings descrevendo as posições em relação ao novo ponto de referência
print('Posição do ponto O em relação ao ponto de referência é: ({0} m,{1} m)'.format(x0R, y0R))
print('Posição do ponto A em relação ao ponto de referência é: ({0} m,{1} m)'.format(xAR, yAR))
print('Posição do ponto B em relação ao ponto de referência é: ({0} m,{1} m)'.format(xBR, yBR))
print('Posição do ponto C em relação ao ponto de referência é: ({0} m,{1} m)'.format(xCR, yCR))

plt.show()
```

Podemos notar que a posição relativa entre os objetos permaneceu a mesma, mas como mudamos o ponto de referência, a maneira de medir as posições mudou - os vetores posição mudaram! Isso quer dizer que a posição de um objeto depende do referencial utilizado!

A posição é uma grandeza vetorial. Como vimos, não é possível definir a posição de um objeto no plano, utilizando apenas um único valor. Para representar um vetor no plano bidimensional, como vimos nos exemplos acima, precisamos de pelo menos dois valores, e com esses dois valores definimos o comprimento do vetor, sua direção (reta) e seu sentido. A posição $\vec{s}$ não é a única grandeza vetorial que será utilizada na Física, logo veremos que a velocidade de um objeto também é uma grandeza vetorial, mas há várias outras. Grandezas vetoriais são sempre representadas com uma seta acima do símbolo ou letra que a representa!

No mundo real, onde temos o espaço tridimensional, é necessário 3 valores diferentes para se definir a posição de qualquer objeto.

```python
%matplotlib inline
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

plt.rcParams["figure.figsize"] = (15,12)

#definindo os pontos no espaço 3D
Ox, Oy, Oz = 0,0,0
Ax, Ay, Az = 8,2,7
Bx, By, Bz = -2,-8,6
Cx, Cy, Cz = -3,6,-9

#desenhando os pontos no espaço 3D, e os vetores que representam as suas posições.
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.scatter(Ox,Oy,Oz, marker = 'o', s = 120)
ax.scatter(Ax,Ay,Az, marker ='*', s = 120)
ax.quiver(Ox,Oy,Oz,Ax,0,0, lw = 4, color = 'orange', arrow_length_ratio = 0.2)
ax.quiver(Ax,Oy,Oz,0,Ay,0, lw = 4, color = 'orange', arrow_length_ratio = 0.2)
ax.quiver(Ax,Ay,Oz,0,0,Az, lw = 4, color = 'orange', arrow_length_ratio = 0.2)
ax.text(Ax - 0.3,Ay,Az + 0.6,'({0} m, {1} m, {2} m)'.format(Ax,Ay,Az), fontsize = 12)
ax.scatter(Bx,By,Bz, marker = 's', s = 120)
ax.quiver(Ox,Oy,Oz,Bx,0,0, lw = 4, arrow_length_ratio = 0.2, color = 'lightgreen')
ax.quiver(Bx,Oy,Oz,0,By,0, lw = 4, arrow_length_ratio = 0.2, color = 'lightgreen')
ax.quiver(Bx,By,Oz,0,0,Bz, lw = 4, arrow_length_ratio = 0.2, color = 'lightgreen')
ax.text(Bx - 0.3,By,Bz + 0.6,'({0} m, {1} m, {2} m)'.format(Bx,By,Bz), fontsize = 12)
ax.scatter(Cx,Cy,Cz, marker = 'P', s = 120)
ax.quiver(Ox,Oy,Oz,0,Cy,0, lw = 4, arrow_length_ratio = 0.2, color = 'pink')
ax.quiver(Ox,Cy,Oz,Cx,0,0, lw = 4, arrow_length_ratio = 0.2, color = 'pink')
ax.quiver(Cx,Cy,Oz,0,0,Cz, lw = 4, arrow_length_ratio = 0.2, color = 'pink')
ax.text(Cx - 0.3,Cy,Cz - 1.2,'({0} m, {1} m, {2} m)'.format(Cx,Cy,Cz), fontsize = 12)

ax.set_xlim(-10,10)
ax.set_ylim(-10,10)
ax.set_zlim(-10,10)
ax.set_xlabel('Posição X (metros)', fontsize = 15)
ax.set_ylabel('Posição Y (metros)', fontsize = 15)
ax.set_zlabel('Posição Z (metros)', fontsize = 15)

plt.show()
```

Um objeto está em movimento em relação a outro objeto, quando sua posição em relação a esse objeto varia a medida que o tempo passa. A grandeza que representa a variação da posição com o tempo é chamada de velocidade $\vec{v}$, e este será nosso próximo assunto.
